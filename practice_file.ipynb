```python
#Check if a string a palindrome or not.
n = "madam"
if n == n[::-1]:
  print('Yes, It is a palindrome.')
else:
  print('Not a palindrome.')
```

```python
#Write a function to find the factorial of a number.
def fact(n):
  if n == 1:
    return n
  else:
    return n * fact(n-1)
fact(5)
```

```python
def even_odd(num):
  if num % 2 == 0:
    print('It is an even number.')
  else:
    print('Odd number.')
even_odd(9)
```

```python
#Write a function to check whether the number is prime or not.
def check_prime(n):
  if n == 1:
    print('Not Prime')
  if n >= 2:
    for i in range(2,n):
      if n % i == 0:
        print('Not a prime number.')
        break
    else:
      print('Prime number.')
check_prime(9)
```
