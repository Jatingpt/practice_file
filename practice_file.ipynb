```python
#How to find the unique and duplicate elements inside a list.
l = [1,2,1,2,3,4,5,6]
unique = []
duplicate = []
for i in l:
  if l.count(i) == 1:
    unique.append(i)
  else:
    duplicate.append(i)
print(unique)
print(list(set(duplicate)))
```
